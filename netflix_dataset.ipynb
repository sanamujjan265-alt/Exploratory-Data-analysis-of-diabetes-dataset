# **Distribution of Type of Netflix Dataset**

## **Distribution of Release Year**

```python
sns.histplot(x=df['release_year'], bins=20, kde=True)
plt.grid()

title_dict = {
    'family': 'Algerian',
    'weight': 'bold',
    'size': 20,
    'color': 'darkred'
}

label_dict = {
    'family': 'Algerian',
    'weight': 'bold',
    'size': 15,
    'color': 'darkred'
}

plt.title("Distribution of Release year", fontdict=title_dict)
plt.xlabel("Release Year", fontdict=label_dict)
plt.ylabel("Count of Release Year", fontdict=label_dict)
plt.show()
```

## **Insight**
- The distribution of release years in the Netflix dataset shows that most titles were released after 2000, with a significant increase in content from 2015 onwards. Older titles (before 2000) are fewer, indicating that Netflix mostly streams modern content while still including some classics.

## **Distribution of Rating**

```python
sns.histplot(x=df['rating'], bins=10, kde=True)
plt.grid()

title_dict = {
    'family': 'Algerian',
    'weight': 'bold',
    'size': 20,
    'color': 'darkred'
}

label_dict = {
    'family': 'Algerian',
    'weight': 'bold',
    'size': 15,
    'color': 'darkred'
}

plt.title("Distribution of Rating", fontdict=title_dict)
plt.xlabel("Rating of Netflix Dataset", fontdict=label_dict)
plt.ylabel("Count of Rating", fontdict=label_dict)
plt.show()
```

## **Insight**
- The distribution of ratings in the Netflix dataset shows that most titles are rated TV-MA and TV-14, indicating content aimed at mature audiences. Fewer titles have ratings like PG or G, showing that kid-friendly content is relatively limited.

## **Count Of Type (Movie & Tv Show) Column**

```python
sns.countplot(x=df['type'])
plt.grid()

title_dict = {
    'family': 'Algerian',
    'weight': 'bold',
    'size': 20,
    'color': 'darkred'
}

label_dict = {
    'family': 'Algerian',
    'weight': 'bold',
    'size': 15,
    'color': 'darkred'
}

plt.title("Count of Movies and TV Shows", fontdict=title_dict)
plt.xlabel("Type", fontdict=label_dict)
plt.ylabel("Count of Type", fontdict=label_dict)
plt.show()
```

## **Insight**
- In the Netflix dataset, the type column shows that Movies are more frequent than TV Shows. This indicates that Netflix’s library has a larger proportion of movies compared to series.

## **Percentage Of Type Column**

```python
x = df['type'].value_counts()

title_dict = {
    "family": 'Algerian',
    'weight': 'bold',
    'size': 20,
    'color': 'darkred'
}

plt.pie(
    x.values,
    labels=x.index,
    autopct='%2.0f%%'
)

plt.title("Percentage of Type Column", fontdict=title_dict)
plt.show()
```

## **Insight**
- In the Netflix dataset, the type column shows that 70% of titles are Movies and 30% are TV Shows, indicating that movies dominate the content library.
