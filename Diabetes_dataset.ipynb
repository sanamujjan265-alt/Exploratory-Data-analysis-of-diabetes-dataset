```python
plt.boxplot(df['Age'].dropna(),vert=False)
plt.show()
```

## **Remove Duplicates**

```python
x=df.drop_duplicates()
```

## **After Remove Duplicates**

```python
def remove_outliers(column):
    Q1=df[column].quantile(0.25)
    Q3=df[column].quantile(0.75)
    IQR=Q3-Q1
    lower_bound = Q1-1.5*IQR
    upper_bound = Q3+1.5*IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
df=remove_outliers('Age')
```

```python
plt.boxplot(df['Age'].dropna(),vert=False)
plt.show()
```

## **Insight**
- missing and incorrect zero value were handeled during data cleaning.
- the dataset was prepared for reliable analysis and prediction.

```python
plt.boxplot(df['Glucose'],vert=False)
plt.title("Boxplot of glucose")
plt.show()
```
